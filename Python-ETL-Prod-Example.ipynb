## **Import** | Packages

```python
import pandas as pd
import requests
from datetime import datetime

check_code_block_execution_status()
```

## **Define** | Variables

```python
path_to_csv_file_to_import = 'Racks.csv'

api_endpoint_get = 'http://192.0.0.0:3005/...'

api_endpoint_post = 'http://192.0.0.0:3005/...'

check_code_block_execution_status()
```

## **Define** | Functions

```python
def post_payload_to_api(payload, api_endpoint):
    response = requests.post(api_endpoint, json = payload)
    print(str(payload['PartNo']) + ' ~ ' + response.text)

check_code_block_execution_status()
```

## **Get** | Api | **Data**

```python
PARAMS = {}
data_ingested_from_api = requests.get(url = api_endpoint_get, params = PARAMS)
data_ingested_from_api = data_ingested_from_api.json()

data_ingested_from_api[0]
```

```python
df_data_ingested_from_api = pd.json_normalize(data_ingested_from_api)
df_data_ingested_from_api.head(1)
```

## **Import** | CSV | **Data**

```python
df_imported_csv_data = pd.read_csv(path_to_csv_file_to_import)

df_imported_csv_data.head(10)
```

## **Set** | Variables

```python
df_to_set_vars = df_imported_csv_data

df_to_set_vars.head(10)
```

```python
where_col_in_quest_is_not_null = pd.notnull(df_to_set_vars['Plant'])
var_val_for_plant = df_to_set_vars[where_col_in_quest_is_not_null]['Plant'].item()

var_val_for_plant
```

```python
where_col_in_quest_is_not_null = pd.notnull(df_to_set_vars['Reason'])
var_val_for_reason = df_to_set_vars[where_col_in_quest_is_not_null]['Reason'].item()

var_val_for_reason
```

```python
val_current_date_time = datetime.now()

str(val_current_date_time)
```

## **Set** | Cageory | **Values**

```python
list_of_df_columns = [
    'Category', 
    'Add'
]

series_new_category_vals = df_to_set_vars[list_of_df_columns].dropna()
series_new_category_vals.set_index("Category", inplace = True)

series_new_category_vals
```

```python
dict_new_category_vals = {}

new_val_for_material_handlers = series_new_category_vals.loc['Material'].item()
new_val_for_test_debug = series_new_category_vals.loc['TestDebug'].item()  
new_val_for_boxing = series_new_category_vals.loc['Boxing'].item()  
new_val_for_tkt_time = series_new_category_vals.loc['TktTime'].item()  
new_val_for_asmblrs = series_new_category_vals.loc['Asmblrs'].item()  
new_val_for_lead = series_new_category_vals.loc['Lead'].item()  
new_val_for_qa = series_new_category_vals.loc['QA'].item()

dict_material_handlers = {'Material': new_val_for_material_handlers}
dict_test_debug = {'TestDebug': new_val_for_test_debug}
dict_boxing = {'Boxing': new_val_for_boxing}
dict_tkt_time = {'TktTime': new_val_for_tkt_time}
dict_asmblrs = {'Asmblrs': new_val_for_asmblrs}
dict_lead = {'Lead': new_val_for_lead}
dict_qa = {'QA': new_val_for_qa}

dict_new_category_vals.update(dict_material_handlers)
dict_new_category_vals.update(dict_test_debug)
dict_new_category_vals.update(dict_boxing)
dict_new_category_vals.update(dict_tkt_time)
dict_new_category_vals.update(dict_asmblrs)
dict_new_category_vals.update(dict_lead)
dict_new_category_vals.update(dict_qa)

dict_new_category_vals
```

## **Get** | GPNs | to **Update**

```python
list_of_df_columns = ['GPN']
list_of_refac_column_names = {'GPN': 'PartNo'}

series_gpns_to_update = df_imported_csv_data[list_of_df_columns] \
                        .dropna() \
                        .astype('int64') \
                        .astype(str) \
                        .rename(columns = list_of_refac_column_names)

series_gpns_to_update.head()
```

```python
val_len_of_gpns_to_update = len(series_gpns_to_update)

val_len_of_gpns_to_update
```

```python
val_gpn_to_update = series_gpns_to_update.iloc[0].item()

val_gpn_to_update
```

```python
for i in range(val_len_of_gpns_to_update):
    print('part_to_be_processed'  + ' | ' + series_gpns_to_update.iloc[i].item())
```

## **Update** | DF

```python
series_gpns_to_update.head(1)
```

```python
df_data_ingested_from_api.head(1)
```

```python
df_gpn_to_update = pd.merge(
    df_data_ingested_from_api, 
    series_gpns_to_update, 
    how='right', 
    on=['PartNo']
)

df_gpn_to_update.head()
```

```python
df_updated_gpn = df_gpn_to_update.copy()

df_updated_gpn.reset_index(inplace = True)
df_updated_gpn.drop(['index'], axis = 1, inplace = True)

df_updated_gpn
```

```python
df_updated_gpn['Material'] = df_updated_gpn['Material'].apply(lambda x: x + new_val_for_material_handlers)
df_updated_gpn['TestDebug'] = df_updated_gpn['TestDebug'].apply(lambda x: x + new_val_for_test_debug)
df_updated_gpn['Boxing'] = df_updated_gpn['Boxing'].apply(lambda x: x + new_val_for_boxing)
df_updated_gpn['TaktTime'] = df_updated_gpn['TaktTime'].apply(lambda x: x + new_val_for_tkt_time)
df_updated_gpn['Assemblers'] = df_updated_gpn['Assemblers'].apply(lambda x: x + new_val_for_asmblrs)
df_updated_gpn['LEAD'] = df_updated_gpn['LEAD'].apply(lambda x: x + new_val_for_lead)
df_updated_gpn['QA'] = df_updated_gpn['QA'].apply(lambda x: x + new_val_for_qa)

df_updated_gpn['ChangeReqLastModDate'] = df_updated_gpn['ChangeReqLastModDate'] = str(val_current_date_time)

check_code_block_execution_status()
```

```python
list_of_df_columns = [
    'Plant',
    'PartNo', 
    'TaktTime',	
    'Assemblers',	
    'Material',	
    'TestDebug',
    'QA',
    'Boxing',
    'LEAD',
    'ChangeReqLastModDate'
]

list_of_dfs = [
    df_gpn_to_update[list_of_df_columns],
    df_updated_gpn[list_of_df_columns],
]

df_to_validate_results = pd.concat(list_of_dfs)

df_to_validate_results
df_to_validate_results.query('PartNo == "1110415"')
```
